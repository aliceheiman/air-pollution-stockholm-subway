# Mapping data by date

```python
# HELPER TIME FORMATTING FUNCTIONS
def add_secs(initial, secs):
    return initial + timedelta(seconds=secs)

def format_time(time_int):
    # convert int to str
    str_time = str(time_int)
    
    # Pad 5:00:00 to 05:00:00
    if len(str_time) < 6:
        str_time = '0' + str_time
    
    # Concatenate date and time
    str_time = date + ' ' + str_time
    
    # Convert string into date time object
    result = datetime.strptime(str_time, '%Y-%m-%d %H%M%S')
    
    return result
```

```python
date = '2021-10-12'
daniel = 'daniel-AM'

### Load DiSC Mini data file
filepath = f'by-date/{date}/{daniel}.txt'

disc_df = pd.read_csv(filepath, sep="\t", skiprows=5)
disc_df = disc_df.iloc[:, :-1] # remove last column
disc_df['Time'] = disc_df['Time'].apply(lambda x: float(x.replace(',', '.'))) # use . as decimal delimiter

# Get start time
with open(filepath) as f:
    contents = f.readlines()
    
match = re.search(r'\d{2}:\d{2}:\d{2}', contents[3])
start_time = match.group()

# Add timestamp column
str_time = date + ' ' + start_time
initial_timestamp = datetime.strptime(str_time, '%Y-%m-%d %H:%M:%S')

disc_df['Timestamps'] = disc_df['Time'].apply(lambda x: add_secs(initial_timestamp, x))
```

```python
### Load Open Seneca
open_df = pd.read_csv(f'by-date/{date}/1-AM.csv', skiprows=1)

# Add timestamp column and offset by two hours to get swedish time
open_df['Timestamps'] = open_df['Time'].apply(lambda x: format_time(x) + timedelta(hours=2))
```

```python
# Plot dataframes
fig, (ax1, ax2) = plt.subplots(dpi=100, figsize = (12, 12), nrows=2, sharex=True)

ax1.plot(open_df['Timestamps'], open_df['PM2.5'], label='Sensor 1')

ax1.set_title(date, fontsize=25)
ax1.set_xlabel('Time')
ax1.set_ylabel('PM 2.5')

ax1.legend(fontsize=20)

ax2.set_ylabel('Particle Count')
ax2.plot(disc_df['Timestamps'], disc_df['Number'], label='Daniel', color='green')
ax2.legend(fontsize=20)

# Format x-axis as 05:34:53 etc.
plt.gcf().autofmt_xdate()
myFmt = dates.DateFormatter('%H:%M:%S')
plt.gca().xaxis.set_major_formatter(myFmt)

plt.show()
```

```python
date = '2021-10-12'

def get_timestamp(timestr, date, offset=0):
    return datetime.strptime(f'{date} {timestr}', '%Y-%m-%d %H:%M:%S') + timedelta(seconds=offset)
```

# START STOP

```python
datafile = './input/Firebase/firebase.json'

with open(datafile, 'r') as f:
    data = json.load(f)
    
route1 = ['Gullmarsplan', 'Skanstull', 'Medborgarplatsen', 'Slussen', 'Gamla Stan', 'T-Centralen', 'Hötorget', 'Rådmansgatan']

# Convert into datetime structure with [date, period, station, start, stop]
structure = {}

# Get data for current date
for station in data['stations'].keys():
    
    # Get timestamps
    for date_time in data['stations'][station].keys():
        
        # Split datetime into date and time
        s_date, s_time = date_time.split(' ')
        
        # Get period (AM or PM)
        period = 'AM' if datetime.strptime(s_time, '%H:%M:%S').hour < 12 else 'PM'
        
        # get route
        route = 'Alice' if station in route1 else 'Alicia'
        
        # Add structure skeleton
        if s_date not in structure:
            structure[s_date] = {}
            
        if station not in structure[s_date]:
            structure[s_date][station] = {}
        
        if period not in structure[s_date][station]:
            structure[s_date][station][period] = {'start': None, 'stop': None, 'route': route, 'mode': None}
        
        # Label action as start or end
        action = data['stations'][station][date_time]['action']
        action = 'start' if action == 'On plattform' else 'stop'
        
        structure[s_date][station][period][action] = s_time
```

```python
# Convert structure into datarows
columns = ['date', 'period', 'station', 'start', 'stop', 'route', 'mode']
rows = []

for date in structure:
    for station in structure[date]:
        for period in structure[date][station]:
            d = structure[date][station][period]
            
            rows.append([date, period, station, d['start'], d['stop'], d['route'], d['mode']])

np_array = np.array(rows)

place_df = pd.DataFrame(data=np_array, columns=columns)

place_df[place_df['date'] == '2021-10-11']

# Save place_df to file
#place_df.to_csv(r'./placetime/placedf1.csv', index=False)
```
